# Tokenize the Tweets

```python
from gensim.models import Word2Vec
```

```python
import pandas as pd
```

```python
df = pd.read_csv('base_elon_data.csv')
```

```python
df.head()
```

```python
tweets = df['tweet'].tolist()
```

```python
tweets
```

```python
df['tweet'] = df['tweet'].astype(str)
```

```python
df['tweet'][0]
```

```python
tweet_list = []
one_tweet = []
for i in range(len(df['tweet'])):
    tweet_list.append(df['tweet'][i].split(' '))#tweets[i].split(' ')
```

```python
tweet_list
```

```python
tweets[0].split(' ')
# be careful with spaces
```

```python
model = Word2Vec(tweet_list, min_count=1)
```
